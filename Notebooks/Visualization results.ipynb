```python
%matplotlib inline
import Orange
import matplotlib.pyplot as plt
```

```python
# Branch length radial visualization heuristic
method_names = ['[BRANCH-LENGTH]-unordered_tree_stress', '[BRANCH-LENGTH]-ordered_tree_stress', '[BRANCH-LENGTH]-ordered_tree_fixed_angle_correction_stress', '[BRANCH-LENGTH]-ordered_tree_adj_based_corrections_stress']
method_average_ranks =  [2.74468085106383, 2.1702127659574466, 2.0851063829787235, 3.0]

cd = Orange.evaluation.compute_CD(method_average_ranks, 48) 
Orange.evaluation.graph_ranks(method_average_ranks, method_names, cd=cd, width=6, textspace=1.5)
plt.show()
```
