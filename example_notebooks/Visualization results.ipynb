```python
# Branch length radial visualization heuristic
method_names = ['[BRANCH-LENGTH]-unordered_tree', '[BRANCH-LENGTH]-ordered_tree', '[BRANCH-LENGTH]-FFAC', '[BRANCH-LENGTH]-ANBC']
method_average_ranks = [2.758807588075881, 3.2357723577235773, 1.8319783197831978, 2.265582655826558] #all leaf nodes
cd = Orange.evaluation.compute_CD(method_average_ranks, 369) 
Orange.evaluation.graph_ranks(method_average_ranks, method_names, cd=cd, width=6, textspace=1.5)
plt.show()
```
